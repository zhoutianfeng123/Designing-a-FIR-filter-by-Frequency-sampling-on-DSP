# calculate the h(n)

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy import signal
import cmath
```

```python
#define the H(k)
def H(k):
    h = 39/40
    N = 40
    if k>=0 and k<7:
        return np.exp(-1j*h*k*np.pi)
    if k==7:
        return 0.388*np.exp(-1j*h*7*np.pi)
    if k>=8 and k<=N-8:
        return 0
    if k==33:
        return 0.388*np.exp(-1j*h*7*np.pi)
    if k>=34 and k<=N-1:
        return np.exp(1j*h*(N-k)*np.pi)
    else:
        print("ERROR!")
```

```python
k = range(0,40,1)
```

```python
n = range(0,40,1)
```

```python
#calculate H(k)
H_k = []
for i in range(40):
    H_k.append(H(i))
```

```python
#calculate |H(k)|
H_abs = []
for i in range(40):
    H_abs.append(abs(H_k[i]))
```

```python
#calculate h(n)
h_n = [0]*40
for i in range(40):#i->n
    m = 0
    for j in range(40):#j->k
        m += H(j)*np.exp(1j*2*np.pi/40*j*i)/40
    h_n[i] = m
```

```python
#calculate the real of h(n)
h_real = []
for i in range(40):
    h_real.append(h_n[i].real)
```

```python
#show
plt.style.use('bmh')
f = plt.figure()
ax1 = f.add_subplot(2,1,1)
ax1.stem(k,H_abs)
ax2 = f.add_subplot(2,1,2)
ax2.stem(n,h_real)
ax1.set_title('abs(H(k))')
ax2.set_title('h(n)')
f.subplots_adjust(hspace=0.5)
plt.show()
```
