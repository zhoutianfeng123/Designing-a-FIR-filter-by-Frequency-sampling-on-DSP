# calculate the h(n)

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy import signal
import cmath
```

```python
#define the H(k)
def H(k):
    h = 19/20
    N = 20
    if k>=0 and k<=3:
        return np.exp(-1j*h*k*np.pi)
    if k>=4 and k<=16:
        return 0
    if k>=17 and k<=19:
        return np.exp(1j*h*(N-k)*np.pi)
    else:
        print("ERROR!")
```

```python
k = range(0,20,1)
```

```python
n = range(0,20,1)
```

```python
#calculate H(k)
H_k = []
for i in range(20):
    H_k.append(H(i))
```

```python
#calculate |H(k)|
H_abs = []
for i in range(20):
    H_abs.append(abs(H_k[i]))
```

```python
#calculate h(n)
h_n = [0]*20
for i in range(20):#i->n
    m = 0
    for j in range(20):#j->k
        m += H(j)*np.exp(1j*2*np.pi/20*j*i)/20
    h_n[i] = m
```

```python
#calculate the real of h(n)
h_real = []
for i in range(20):
    h_real.append(h_n[i].real)
```

```python
#show
plt.style.use('bmh')
f = plt.figure()
ax1 = f.add_subplot(2,1,1)
ax1.stem(k,H_abs)
ax2 = f.add_subplot(2,1,2)
ax2.stem(n,h_real)
ax1.set_title('abs(H(k))')
ax2.set_title('h(n)')
f.subplots_adjust(hspace=0.5)
plt.show()
```
